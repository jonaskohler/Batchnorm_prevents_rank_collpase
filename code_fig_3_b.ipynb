```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d=50
L=100

evs=[]

gamma=0.01
M = np.eye(d)
x=np.random.randn(d)
H=x@x.T+0.00001*np.random.randn(d,d)
M=H@H.T
I = np.eye(d) 
evs=[]
for k in range(L):
    W=np.random.randn(d,d)*100
    Wh =  I + gamma*W
    Mh = Wh @ M @ Wh.T
    DM = np.diag(1./np.sqrt(np.diag(Mh)))
    M = DM @ Mh @ DM

    evs.append(np.linalg.eigh(M)[0])
       
evs=np.array(evs)     
```

```python
plt.plot(evs[:,-1],label="$\lambda_1$")
plt.plot(evs[:,-2],label="$\lambda_2$")
plt.plot(evs[:,-3],label="$\lambda_3$")
plt.plot(evs[:,-4],label="$\lambda_4$")
plt.plot(evs[:,-5],label="$\lambda_5$")
plt.plot(evs[:,-6],label="$\lambda_6$")
plt.plot(evs[:,-7],label="$\lambda_7$")
plt.plot(evs[:,-8],label="$\lambda_8$")
plt.plot(evs[:,-9],label="$\lambda_9$")
plt.plot(evs[:,-10],label="$\lambda_{10}$")


plt.title("First ten eigenvalues of $M_\ell$")

plt.xlabel("hidden layer ($\ell$)",fontsize=12)
plt.yscale('log')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 0.25),fancybox=True, shadow=True, ncol=5,fontsize=9)

#plt.savefig(evs.pdf)
```
